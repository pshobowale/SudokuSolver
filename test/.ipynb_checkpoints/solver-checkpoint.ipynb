```python

def getSolved(S):
    R=np.zeros((9,9))
    
    for x in range(9):
        for y in range(9):
            sum=0
            v=0
            for i in range(9):
                sum+=S[x,y,i]
                if(S[x,y,i]):
                    v=int(i)
            if sum==1:
                R[x,y]=v+1
            if sum==0:
                print("Error no Solution for Field: ",x,",",y)
    
    return R

   
```

```python
def getClue(S):
    S=np.copy(S)
    R=getSolved(S)
    
    for x in range(9):
        for y in range(9):
            v=R[x,y]
            if v!=0:
                #print(v,R[x,:])
                S[x,:,int(v-1)]=0
                #print(v,R[:,y])
                S[:,y,int(v-1)]=0
                
                
                xk,yk=x//3,y//3
                #print(x,y,"|",xk,yk)
                #print(R[xk*3:xk*3+3,yk*3:yk*3+3])
                S[xk*3:xk*3+3,yk*3:yk*3+3,int(v-1)]=0
                S[x,y,int(v-1)]=1
    return S
                
def printSolved(S):
    if(len(S.shape)==3):
        R=getSolved(S)
    else:
        R=S
    for y in range(9):
        for x in range(9):
            if(R[x,y]):
                print(int(R[x,y]),"\t",end='')
            else:
                print(" \t",end='')
            if(x%3==2):
                print("|",end='')
        
        print()
        if(y%3==2):
            print("-\t-\t-\t-\t-\t-\t-\t-\t")
            
        
    
    print("---------------------------------------------------------------------------\n\n\n") 

printSolved(S)
```

```python
def printDifference(A,B):
    printSolved(abs(getSolved(A)-getSolved(B)))
    
def Change(A,B):
    return(np.sum(A-B))
```

```python

def printPossibilitys(S):
    R=np.zeros((9,9))
    
    for x in range(9):
        for y in range(9):
            sum=0
            for i in range(9):
                sum+=S[x,y,i]
            R[x,y]=sum
    printSolved(R)
```

```python
Org=np.copy(S)
printSolved(S)
New=getClue(S)
printDifference(S,New)
while(Change(S,New)):
    S=New
    New=getClue(S)
    print("Change",Change(S,New))
    printDifference(S,New)

print("Result")
printSolved(S)
print("Difference to Original")
printDifference(S,Org)
print("Possibilitys")
printPossibilitys(S)
```
